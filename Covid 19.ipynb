```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
pat_df= pd.read_csv(r'C:\Users\HP\Downloads\covid_19_india.csv')
#pat_df= pat_df.iloc[::-1] #making dates in ascending order
dates = pat_df['Date'].tolist()
x = list( dict.fromkeys(dates) ) # removing duplicates
y= pat_df['Date'].value_counts().tolist()
y=y[::-1] # same as above

plt.plot(x,y)

plt.xlabel("Date")
plt.ylabel("Covid positive  per day")
plt.xticks(np.arange(0, 60, 10))
plt.yticks(np.arange(0,30,2))
plt.title("Covid-19 Spread Rate (till 21/03/2020)")
plt.savefig('covid-19.png')
```
